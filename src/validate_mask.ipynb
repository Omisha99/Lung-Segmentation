```python
import matplotlib.pyplot as plt
from PIL import Image

img = Image.open("../raw_data/shcxr/images/CHNCXR_0001_0.png")
mask = Image.open("../raw_data/shcxr/masks/CHNCXR_0001_0_mask.png")

plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.imshow(img, cmap="gray")
plt.title("Image")

plt.subplot(1,3,2)
plt.imshow(mask, cmap="gray")
plt.title("Mask")

plt.subplot(1,3,3)
plt.imshow(img, cmap="gray")
plt.imshow(mask, cmap="jet", alpha=0.3)
plt.title("Overlay")

plt.show()
```
